# Step  1: Import Libraries

```python
%matplotlib inline

import os
import numpy as np
import tensorflow as tf

from PIL import Image
from matplotlib import pyplot as plt
```

# Step 2: Rotation

```python
generator = tf.keras.preprocessing.image.ImageDataGenerator(
    rotation_range=40
)
```

```python
image_path = 'images/train/cat/cat.jpg'

plt.imshow(plt.imread(image_path));
```

```python
x, y = next(generator.flow_from_directory('images', batch_size=1))
plt.imshow(x[0].astype('uint8'));
```

# Step 3: Width and Height Shifts

```python
generator = tf.keras.preprocessing.image.ImageDataGenerator(
   width_shift_range=[-90 , -40 , 0, 40, 90],
   height_shift_range=[-40 , 0, 40]
)
```

```python
x, y = next(generator.flow_from_directory('images', batch_size=1))
plt.imshow(x[0].astype('uint8'));
```

# Step 4: Brightness

```python
generator = tf.keras.preprocessing.image.ImageDataGenerator(
    brightness_range=(0.5,2.)
)

x, y = next(generator.flow_from_directory('images', batch_size=1))
plt.imshow(x[0].astype('uint8'));
```

# Step 5: Shear Transformation

```python
generator = tf.keras.preprocessing.image.ImageDataGenerator(
    shear_range=40
)

x, y = next(generator.flow_from_directory('images', batch_size=1))
plt.imshow(x[0].astype('uint8'));
```

# Step 6: Zoom

```python
generator = tf.keras.preprocessing.image.ImageDataGenerator(
   zoom_range=0.5
)

x, y = next(generator.flow_from_directory('images', batch_size=1))
plt.imshow(x[0].astype('uint8'));
```

# Step 7: Channel Shift

```python
generator = tf.keras.preprocessing.image.ImageDataGenerator(
    channel_shift_range=100
)

x, y = next(generator.flow_from_directory('images', batch_size=1))
plt.imshow(x[0].astype('uint8'));
```

# Step 8: Flips

```python
generator = tf.keras.preprocessing.image.ImageDataGenerator(
   horizontal_flip=True,
    vertical_flip=True,
    rotation_range=40
)

x, y = next(generator.flow_from_directory('images', batch_size=1))
plt.imshow(x[0].astype('uint8'));
```

# Step 9: Normalization

### Featurewise

```python
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.cifar10.load_data()

generator = tf.keras.preprocessing.image.ImageDataGenerator(
    featurewise_center=True,
    featurewise_std_normalization=True
)

generator.fit(x_train)
```

```python
x, y = next(generator.flow(x_train, y_train, batch_size=1))
print(x.mean(), x.std(), y)
print(x_train.mean())
```

### Samplewise

```python
generator = tf.keras.preprocessing.image.ImageDataGenerator(
   samplewise_center=True,
   samplewise_std_normalization=True
)

x, y = next(generator.flow(x_train, y_train, batch_size=1))
print(x.mean(), x.std(), y)
```

# Step 10: Rescale and Preprocessing Function

```python
generator = tf.keras.preprocessing.image.ImageDataGenerator(
   rescale=1,
    preprocessing_function=tf.keras.applications.mobilenet_v2.preprocess_input
)
```

```python
x, y = next(generator.flow(x_train, y_train, batch_size=1))
```

```python
print(x.mean(), x.std(), y)
```

# Step 11: Using in Model Training

# The code here

```python
model=tf.keras.models.Sequential([tf.keras.applications.mobilenet_v2.MobileNetV2(include_top=False,input_shape=(32,32,3),pooling='max'),tf.keras.layers.Dense(
10,activation='softmax')])
model.compile(loss='sparse_categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
_= model.fit(generator.flow(x_train, y_train, batch_size=32),epochs=1,steps_per_epoch=10 )
```
